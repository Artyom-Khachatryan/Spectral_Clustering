```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r'C:\Users\User\Downloads\hospital_deaths_train.csv')
data = data.fillna(np.mean(data, axis = 0))
data = data.drop('recordid', axis = 1)
X = data.drop('In-hospital_death', axis = 1)
y = data['In-hospital_death']
x_train, x_test, y_train, y_test = train_test_split(X, y,random_state=0, test_size=0.2)
data
```

```python
model = tf.keras.models.Sequential([
  #tf.keras.layers.Flatten(input_shape=(2600,114)),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(128, activation = 'relu'),
  tf.keras.layers.Dense(32, activation = 'sigmoid')
  tf.keras.layers.Dense(128, activation = 'tanh'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(2)
])
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=20)
```

```python
d = pd.read_csv(r'C:\Users\User\Downloads\hospital_deaths_test.csv')
d = d.fillna(np.mean(data, axis = 0))
d = d.drop('recordid', axis = 1)
X = d.drop('In-hospital_death', axis = 1)
y = d['In-hospital_death']
```

```python
model.evaluate(X,  y, verbose=2)
```
