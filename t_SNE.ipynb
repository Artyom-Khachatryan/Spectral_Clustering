```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
class tsne:
    def __init__(self, n_components: int = 2, learning_rate: float = 0.001, n_iter: int = 50, perplexity = 30, momentum = 0.5):
        self.n_components = n_components
        self.n_iter = n_iter
        self.learning_rate = learning_rate
        self.perplexity = perplexity
        self.momentum = momentum
        self.sigma = None
                        
    def gradient(self, y):
        k = np.zeros(self.n_componets)
        for j in range(X.shape[0]):
            for i in range(X.shape[0]):
                if i != j:
                    k += ((self.p[i,j]-self.q[i, j])*(y[i]- y[j]))/((1+np.linalg.norm(y[i]-y[j])**2))#, 1/(1+np.linalg.norm(y-y[j])**2))
        
        return k 
    def fit_transform(self,X):
        scaler  = MinMaxScaler()
        scaler.fit(X)
        X= scaler.transform(X)
        
        g_parameter = np.random.normal(0, 0.0001, (X.shape[0], self.n_components))
        y = np.random.normal(0, 0.0001, (X.shape[0], self.n_components))
        
        self.p = np.zeros((X.shape[0], X.shape[0]))
        self.q = np.zeros((X.shape[0], X.shape[0]))
        self.sigma = 5*np.random.rand(X.shape[0])
        H = 0
        while 2**H < self.perplexity:
            for i in range(X.shape[0]):
                for j in range(X.shape[0]):
           H -= self.p[j][i]*np.log2(self.p[j][i])
           self.sigma += 0.1 
        for _ in range(self.n_iter):
            for i in range(X.shape[0]):
                for j in range(X.shape[0]):
                    if i !=j:
                        a = np.exp(-np.linalg.norm(X[i]-X[j])**2)/(2*self.sigma[i])**2
                        b = 0
                        
                        c = 1/(1+np.linalg.norm(y[i]-y[j])**2)
                        d = 0
                        for g in range(X.shape[0]):
                            if g!=i:
                                b+= np.exp(-np.linalg.norm(X[i] - X[g])**2)/(2*self.sigma)**2
                                d+= 1/(1+np.linalg.norm(y[i]-y[j])**2)
                        self.p[i][j] = a/b
                        self.q[i][j] = c/d
                g_parameter = y[i]
                y[i] = y[i] + self.learning_rate * self.gradient(y[i]) + self.momentum * (y[i]-g_parameter[i])
                
                
            
        return y
        
        

model = tsne()
```

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load MNIST dataset
mnist = fetch_openml('mnist_784')

# Split the data into features (X) and labels (y)
X = mnist['data']
y = mnist['target']

# Scale the features to a range of 0 to 1
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Apply t-SNE
tsne = tsne(n_components=2)
X_tsne = tsne.fit(X_scaled)

# Plot the t-SNE visualization
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10')
plt.colorbar()
plt.title("t-SNE Visualization of MNIST")
plt.show()
```

```python
from sklearn.datasets import fetch_openml
import numpy as np
import matplotlib.pyplot as plt
 
# Load the MNIST data
X, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False)
 
# Randomly select 1000 samples for performance reasons
np.random.seed(100)
subsample_idc = np.random.choice(X.shape[0], 1000, replace=False)
X = X[subsample_idc,:]
y = y[subsample_idc]
 
# Show two example images
fig, ax = plt.subplots(1,2)
ax[0].imshow(X[11,:].reshape(28,28), 'Greys')
ax[1].imshow(X[15,:].reshape(28,28), 'Greys')
ax[0].set_title("Label 3")
ax[1].set_title("Label 8")
```

```python
from sklearn.manifold import TSNE
import pandas as pd
import seaborn as sns
 
# We want to get TSNE embedding with 2 dimensions
n_components = 2
tsne = TSNE(n_components)
tsne_result = tsne.fit_transform(X)
tsne_result.shape
# (1000, 2)
# Two dimensions for each of our images
 
# Plot the result of our TSNE with the label color coded
# A lot of the stuff here is about making the plot look pretty and not TSNE
tsne_result_df = pd.DataFrame({'tsne_1': tsne_result[:,0], 'tsne_2': tsne_result[:,1], 'label': y})
fig, ax = plt.subplots(1)
sns.scatterplot(x='tsne_1', y='tsne_2', hue='label', data=tsne_result_df, ax=ax,s=120)
lim = (tsne_result.min()-5, tsne_result.max()+5) 
ax.set_xlim(lim)
ax.set_ylim(lim)
ax.set_aspect('equal')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
```
